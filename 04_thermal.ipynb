```python
from workflows.pyiron.thermodynamics import calculate_thermal_properties
from workflows.pyiron.build import bulk, repeat
from pyiron_workflow import Workflow
from workflows.pyiron.kg import KnowledgeDict 
```

```python
pair_style = "eam/alloy"
pair_coeff = "* * workflows/potentials/Fe_Ack.eam Fe"
```

```python
kg = KnowledgeDict()
```

```python
wf = Workflow("thermal_expansion")
wf.bulk = bulk('Fe', cubic=True, kg=kg)
wf.structure = repeat(wf.bulk, (3,3,3), kg=kg)
wf.thermal = calculate_thermal_properties(wf.structure, 
    pair_style, 
    pair_coeff, 
    temperature=293, pressure=0,
    n_run_steps=10000,
    kg=kg)
```

```python
wf.run()
```

```python
kg.to_yaml('yamls/05_thermal.yaml')
```
