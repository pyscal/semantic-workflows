```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the data, skip comment lines starting with #
df = pd.read_csv("Result.txt", comment='#', header=None, names=['displacement', 'SFE_mJ_per_m2'])

# Plot SFE vs Displacement
plt.figure(figsize=(8, 5))
plt.plot(df['displacement'], df['SFE_mJ_per_m2'], marker='o', linestyle='-')

plt.xlabel('Displacement (box units)')
plt.ylabel('Stacking Fault Energy (mJ/m)')
plt.title('Stacking Fault Energy vs Displacement')
plt.grid(True)
plt.tight_layout()
plt.savefig('sfe_plot.png', dpi=300)
plt.show()
```
